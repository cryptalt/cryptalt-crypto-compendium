# Investing Basics 


## Interest

| Terrm              | Meaning          |
| :-- | :-- |
| Debt               | Money owed -     |
| Asset              | Money owned +    |
| Principal          | The initial amount before any interest |
| Interest           | Percent added to balance of the debt or asset at fixed interval  |
| Compound Interest  | Interest on the new balance, which is higher due to previous interest. |
| APR                | Annual percentage rate of interest excluding compound interest |
| APY                | Annual percentage yield of interest including compound interest. APY is higher than APR |
| rate of supply change | the rate at which the supply of a currency changes. This can be linear, decreasing or dynamic |
| inflation          | The rate of decrease in the purchasing power of a currency. High increases in the supply of a currency typically cause high inflation |

## APR vs APY
The larger the APR, the larger the difference between APR and APY
When comparing rates, be sure you are comparing APR to APR, or APY to APY


| APR  | APY    | 
| :--  | :--    | 
| 10%  | 10.5%  |
| 100% | 171.5% |
| 200% | 635%   |
| 500% | 14,246% |

Site for converting APR <-> APY: https://www.aprtoapy.com/

##### Time compounds exponentially

As you can see in the table below, you can double your final amount either by doubling your initial principal or doubling the APR.

Doubling the duration, however can increase your amount by more than 70x, depending on the APR and duration.

| Principal   | APR   | Years | Final Amount  |
| :--------   | :---- | :-----| :-----------  |
| \\$1000     | 1%    | 25    | \\$1,284      |
| \\$1000     | 1%    | 50    | \\$1,648      |

| \\$1000     | 10%   | 50    | \\$150,000    |
| \\$1000     | 20%   | 25    | \\$300,000    |
| \\$2000     | 10%   | 50    | \\$300,000    |
| \\$1000     | 20%   | 50    | \\$22,000,000 |
| \\$2000     | 20%   | 50    | \\$44,000,000 |


```python


p = 1000
n = 365
t = np.arange(0.0, 21.0, 1.0)

def compound_interest(p, r, n, t):
    return p * (1 + r/n)**(n*t)

def ten_percent(t):
    return compound_interest(p, 10/100, n, t)

def twenty_percent(t):
    return compound_interest(p, 20/100, n, t)

def thirty_percent(t):
    return compound_interest(p, 30/100, n, t)

# Plotting both the curves simultaneously
plt.plot(t, ten_percent(t), color='r', label='10%')
plt.plot(t, twenty_percent(t), color='g', label='20%')
plt.plot(t, thirty_percent(t), color='b', label='30%')
  
# Naming the x-axis, y-axis and the whole graph
plt.xlabel("years")
plt.ylabel("amount")
plt.title("$1000 principal for up to 30% APR")
  
#plt.xlim([0, 20])
# Adding legend, which helps us recognize the curve according to it's color
plt.legend()
  
# To load the display window
plt.show()

state = False
toggle_code(state)
```

### Formulas:

#### Compound Interest
a = p * (1 + r/n)^(nt)


Where a = amount, = principal, r = apr, n = number of compoundings per unit time, t = count of unit time (usually year)

#### APY to APR
const apyToApr = (interest, frequency) => ((1 + (interest / 100)) ** (1 / frequency) - 1) * frequency * 100;

#### APR to APY
const aprToApy = (interest, frequency) => ((1 + (interest / 100) / frequency) ** frequency - 1) * 100;

## Automated strategies for Buying


### Lump Sum Investing

> The best time to buy Bitcoin was 5 years ago. The second best time is now.

With lump sum investing, you invest a large sum and never sell, with the goal of maximizing time in the market.
Lump sum also involves timing the market. Effective lump sum investing requires timing the market. Since crypto prices regularly crash 30% or more, you don't want to go all-in when prices are near an all-time high.


### Dollar Cost Averaging (DCA)
 
With dollar cost averaging, you purchase at a regular interval, e.g. weekly, monthly.
This can be an extremely effective strategy, especially for new investors or those with limited time or capital.
If you are new to investing, or have been losing money trading, strongly consider this strategy.

Advantages of dollar cost averaging
* don't need to time the market
* very little time required
* limits risk 
* trains to "buy the dip" instead of panic sell
* uses volatility to lower cost basis 

Dollar cost averaging may even outperform a lump sum if the lump sum was bought near a new all time high, as in the December 2017 example below.

### Market cycle indicators 
#### Logarithmic Regression - BTC/ETH rainbow chart

In this strategy, you buy when the price of Bitcoin or Ethereum is significantly below it's logarithmic regression trend (blue/green) and you sell when it's significantly above (orange/red). This strategy can be combined with either lump sum or weighted dollar cost averaging, where the amount you buy increases as the price is "cheap" on the regression model. 

There is no guarantee this regression model will hold in the future.

<img src="bitcoin-rainbow.png" width="90%" height="90%" />

Source: [https://www.blockchaincenter.net/bitcoin-rainbow-chart/](https://www.blockchaincenter.net/bitcoin-rainbow-chart/)

See also: [https://www.blockchaincenter.net/ethereum-rainbow-chart/](https://www.blockchaincenter.net/ethereum-rainbow-chart/)

#### Pi Cycle Top Indicator

In this strategy, you wait for the value of the 350 daily moving average x2 (orange line) to pass the 111 daily moving averagee (green line) as a sell signal. This indicator has often timed market tops to within a few days.
